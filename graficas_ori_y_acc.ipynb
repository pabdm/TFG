```python
import numpy as np
import pickle
import matplotlib.pyplot as plt

# Definir una función para leer un archivo pkl
def leer_archivo_pkl(nombre_archivo):
    try:
        with open(nombre_archivo, 'rb') as archivo:
            contenido = pickle.load(archivo)
            return contenido
    except FileNotFoundError:
        print(f"El archivo {nombre_archivo} no se encontró.")
    except Exception as e:
        print(f"Error al leer el archivo {nombre_archivo}: {e}")

# Ruta del archivo pkl
nombre_archivo_pkl = './Pruebas 1/walk_002_stageii_pkl_version_synthetic_poco_ruido_omegas.pkl'

# Llama a la función para leer el archivo pkl
datos = leer_archivo_pkl(nombre_archivo_pkl)

# Extraer velocidades angulares y aceleraciones del diccionario de datos
velocidades_angulares = datos['ori']
aceleraciones = datos['acc']

# Diferencial de tiempo
DeltaT = 1/120

# Listas para almacenar las velocidades angulares y aceleraciones en cada eje
omega_x = [[] for _ in range(8)]
omega_y = [[] for _ in range(8)]
omega_z = [[] for _ in range(8)]
aceleracion_x = [[] for _ in range(8)]
aceleracion_y = [[] for _ in range(8)]
aceleracion_z = [[] for _ in range(8)]

# Iterar sobre los valores de j (suponiendo que son 8 en total)
for j in range(8):
    # Iterar sobre todas las velocidades angulares en cada IMU
    for i in range(len(velocidades_angulares)):
        # Extraer las velocidades angulares en cada eje (x, y, z) de cada IMU
        omega_x[j].append(velocidades_angulares[i][j][0])
        omega_y[j].append(velocidades_angulares[i][j][1])
        omega_z[j].append(velocidades_angulares[i][j][2])
        
        # Extraer las aceleraciones en cada eje (x, y, z) de cada IMU
        aceleracion_x[j].append(aceleraciones[i][j][0])
        aceleracion_y[j].append(aceleraciones[i][j][1])
        aceleracion_z[j].append(aceleraciones[i][j][2])

# Lista de valores de j
valores_j = range(len(omega_x))

# Número de subgráficos (uno para cada valor de j)
num_subplots = len(valores_j)

# Crear subgráficos para velocidades angulares y aceleraciones
fig, axs = plt.subplots(num_subplots, 2, figsize=(12, 8), sharex=True)

# Iterar sobre cada valor de j y graficar en el subgráfico correspondiente
for j, ax in zip(valores_j, axs):
    # Plot para velocidades angulares
    ax[0].plot(omega_x[j], label='omega_x')
    ax[0].plot(omega_y[j], label='omega_y')
    ax[0].plot(omega_z[j], label='omega_z')
    ax[0].set_title('IMU {}'.format(j+1))
    if j == 0:
        ax[0].legend()
    
    # Plot para aceleraciones
    ax[1].plot(aceleracion_x[j], label='aceleracion_x')
    ax[1].plot(aceleracion_y[j], label='aceleracion_y')
    ax[1].plot(aceleracion_z[j], label='aceleracion_z')
    if j == 0:
        ax[1].legend()

# Etiqueta de eje x común para todos los subgráficos
axs[-1, 0].set_xlabel('Índice de tiempo')
axs[-1, 1].set_xlabel('Índice de tiempo')

# Ajustar espaciado entre subgráficos
plt.tight_layout()

# Mostrar la gráfica
plt.show()
```

```python
import numpy as np
import pickle
import matplotlib.pyplot as plt

# Definir una función para leer un archivo pkl
def leer_archivo_pkl(nombre_archivo):
    try:
        with open(nombre_archivo, 'rb') as archivo:
            contenido = pickle.load(archivo)
            return contenido
    except FileNotFoundError:
        print(f"El archivo {nombre_archivo} no se encontró.")
    except Exception as e:
        print(f"Error al leer el archivo {nombre_archivo}: {e}")

# Ruta del archivo pkl
nombre_archivo_pkl = './Pruebas 1/walk_002_stageii_pkl_version_synthetic_poco_ruido_omegas.pkl'

# Llama a la función para leer el archivo pkl
datos = leer_archivo_pkl(nombre_archivo_pkl)

# Convertir aceleraciones a una matriz NumPy
aceleraciones = np.array(datos['acc'])

# Diferencial de tiempo
DeltaT = 1/120

# IMUs de interés (IMU 1 y IMU 8)
imu1 = 0
imu8 = 7

# Extraer aceleraciones para la IMU 1 y la IMU 8
aceleracion_imu1_x = aceleraciones[:len(aceleraciones)//4, imu1, 0]
aceleracion_imu1_y = aceleraciones[:len(aceleraciones)//4, imu1, 1]
aceleracion_imu1_z = aceleraciones[:len(aceleraciones)//4, imu1, 2]

aceleracion_imu8_x = aceleraciones[:len(aceleraciones)//4, imu8, 0]
aceleracion_imu8_y = aceleraciones[:len(aceleraciones)//4, imu8, 1]
aceleracion_imu8_z = aceleraciones[:len(aceleraciones)//4, imu8, 2]

# Calcular la duración total de la muestra en segundos
duracion_segundos = len(aceleraciones) * DeltaT

# Imprimir la duración total de la muestra por la terminal
print(f"Duración total de la muestra: {duracion_segundos} segundos")

# Crear la figura y los ejes
fig, axs = plt.subplots(2, figsize=(10, 10))

# Graficar las aceleraciones de la IMU 1
axs[0].plot(aceleracion_imu1_x, label='aceleracion_x')
axs[0].plot(aceleracion_imu1_y, label='aceleracion_y')
axs[0].plot(aceleracion_imu1_z, label='aceleracion_z')
axs[0].set_title('IMU 1 - Aceleración')
axs[0].legend()

# Graficar las aceleraciones de la IMU 8
axs[1].plot(aceleracion_imu8_x, label='aceleracion_x')
axs[1].plot(aceleracion_imu8_y, label='aceleracion_y')
axs[1].plot(aceleracion_imu8_z, label='aceleracion_z')
axs[1].set_title('IMU 8 - Aceleración')
axs[1].legend()

# Etiqueta de eje x común para ambos subgráficos
axs[1].set_xlabel('Índice de tiempo')

# Ajustar espaciado entre subgráficos
plt.tight_layout()

# Mostrar la gráfica
plt.show()
```
