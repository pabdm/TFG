***Código con quaterniones***

```python
from madgwickahrs import MadgwickAHRS
from quaternion import Quaternion  # Importa la clase Quaternion del segundo archivo
import numpy as np
import pickle
import csv

def leer_archivo_pkl(nombre_archivo):
    try:
        with open(nombre_archivo, 'rb') as archivo:
            contenido = pickle.load(archivo)
            return contenido
    except FileNotFoundError:
        print(f"El archivo {nombre_archivo} no se encontró.")
    except Exception as e:
        print(f"Error al leer el archivo {nombre_archivo}: {e}")

nombre_archivo_pkl = './Pruebas 1/walk_002_stageii_pkl_version_synthetic_omegas.pkl'

# Crear una instancia de MadgwickAHRS
ahrs = MadgwickAHRS()

# Llama a la función para leer el archivo pkl
datos = leer_archivo_pkl(nombre_archivo_pkl)

# Extraer matrices de rotación y aceleraciones del diccionario de datos
gyroscope_data = datos['ori']
accelerometer_data = datos['acc']

# Lista para almacenar los ángulos de Euler para cada IMU en cada instante de tiempo
euler_angles = []

# Lista para almacenar los quaterniones para cada IMU en cada instante de tiempo
quaternions = []

# Iterar sobre cada instante de tiempo
for instant in range(len(gyroscope_data)):
    # Lista para almacenar los ángulos de Euler de cada IMU en este instante de tiempo
    euler_angles_instant = []
    
    # Lista para almacenar los quaterniones de cada IMU en este instante de tiempo
    quaternions_instant = []
    
    # Iterar sobre cada IMU
    for imu in range(len(gyroscope_data[instant])):
        # Crear una nueva instancia de MadgwickAHRS para cada IMU
        ahrs = MadgwickAHRS()
        
        # Actualizar la orientación con los datos del sensor
        ahrs.update_imu(gyroscope_data[instant][imu], accelerometer_data[instant][imu])
        
        # Obtener los ángulos de Euler y agregarlos a la lista
        euler_angle = ahrs.quaternion.to_euler_angles()
        euler_angles_instant.append(np.degrees(euler_angle))  # Convertir a grados
        
        # Agregar el quaternion a la lista
        quaternions_instant.append(ahrs.quaternion)
        
    
    # Agregar los ángulos de Euler de este instante a la lista principal
    euler_angles.append(euler_angles_instant)
    
    # Agregar los quaterniones de este instante a la lista principal
    quaternions.append(quaternions_instant)

# Guarda los ángulos de Euler en un archivo (opcional)
with open('euler_angles.pkl', 'wb') as archivo:
    pickle.dump(euler_angles, archivo)

# Guarda los quaterniones en un archivo (opcional)
with open('./Pruebas 1/walk_002_stageii_pkl_version_synthetic_omegas_quaternions.pkl', 'wb') as archivo:
    pickle.dump(quaternions, archivo)


# Iterar sobre cada instante de tiempo
for instant, instant_quaternions in enumerate(quaternions):
    print(f"Instante de tiempo {instant}:")
    
    # Iterar sobre cada IMU
    for imu, quaternion in enumerate(instant_quaternions):
        print(f"IMU {imu}: {quaternion.q}")
        

nombre_archivo_csv = './Pruebas 1/quaternions_8.csv'

# Guardar los quaterniones de la primera IMU en un archivo CSV
with open(nombre_archivo_csv, 'w', newline='') as archivo_csv:
    writer = csv.writer(archivo_csv)
    for instant_quaternions in quaternions:
        # Tomar solo el primer cuaternión de la primera IMU
        quaternion_first_imu = instant_quaternions[7]
        writer.writerow(quaternion_first_imu.q)
```
